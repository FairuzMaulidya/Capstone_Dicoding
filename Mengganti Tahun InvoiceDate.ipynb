```python
import pandas as pd
import numpy as np

df = pd.read_csv('online_retail.csv')

# Konversi kolom InvoiceDate
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M', errors='coerce')

# Cek hasil konversi
print("Setelah konversi datetime:")
print(df['InvoiceDate'].head())

# Rubah jadi tahun 2024
np.random.seed(42)
random_years = np.random.choice([2024], size=len(df))

# Mengganti bagian tahun
df['InvoiceDate'] = [
    date.replace(year=year) if pd.notnull(date) else pd.NaT
    for date, year in zip(df['InvoiceDate'], random_years)
]

# Menapilkan hasil
print("\nSetelah tahun diganti:")
print(df['InvoiceDate'].head(10))
```

```python
df.to_csv('online_retail1.csv', index=False)
```
